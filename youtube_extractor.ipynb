````python
import youtube_dl
import pandas as pd
import numpy as np
````

````python
pd.set_option("display.max_columns", 65)

options = {
    'simulate': True, # n√£o fazer o download dos videos
    'ignoreerrors': True 
}

youtube = youtube_dl.YoutubeDL(options)
````


````python
queries = ['machine+learning', 'data+science', 'kaggle'] # procurando por videos com estas palavras chaves

results = []
result = None
for query in queries:
    result = youtube.extract_info(f"ytsearchdate2000:{query}") # buscar 2000 registros de cada palavra chave
    for entry in result['entries']:
        if entry is not None:
            entry['query'] = query
    results += result['entries']
````

````python
row_counts1 = len(results)
results = [r for r in results if r is not None] # eliminando registros None
row_counts2 = len(results)
print(f'{row_counts1 - row_counts2} registros None eliminados')
````

````python
df = pd.DataFrame(results)
df['upload_date'] = pd.to_datetime(df['upload_date'])
df['time_since_pub'] = (pd.to_datetime('2020-12-09') - df['upload_date']) / np.timedelta64(1, 'D') # convertendo data para quantidade de dias
print(df.shape)
````

````python
df.drop(labels=['requested_subtitles', 'requested_formats'], axis=1, inplace=True) # estas colunas estavam causando erro no
                                                                                   # momento em que o df era salvo
````

````python
df.to_feather('raw_data.feather')
````
